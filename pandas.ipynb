```python
import pandas as pd

serie= pd.Series([12,58,63,47,81,35])

print(serie)

suma_serie = serie.sum()
print("L’import total és:", suma_serie)

media=serie.mean()
print(round(media,2))

raro = serie.value_counts()
print(raro)

max = serie.max()
print(max)

min = serie.min()
print(min)

raro_dos = serie.cumsum()
print(raro_dos)
```
